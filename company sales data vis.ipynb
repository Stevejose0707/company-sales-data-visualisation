```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/Steve Jose/Desktop/data visualisation/company_sales_data.csv")
df
```

```python
df.columns
```

```python
#Read Total profit of all months and show it using a line plot
plt.Figure(figsize=(10,5))
plt.plot(df['month_number'],df['total_profit'],marker='o',linestyle='-',color='b')
plt.xlabel('month')
plt.ylabel('profit')
plt.title('Monthly Profit')
plt.legend()
plt.grid(axis="x")
plt.show()
```

```python
# 2. Get total profit of all months and show line plot with the following Style properties
#  Generated line plot must include following Style properties:
#  Line Style dotted and Line-color should be red
#  Show legend at the lower right location.
#  X label name = Month Number
#  Y label name = Sold units number
#  Add a circle marker.
#  Line marker color as read
#  Line width should be 3
plt.Figure(figsize=(10,5))
plt.plot(df['month_number'],df['total_profit'],marker='o',linestyle='dotted',color='r',linewidth=3,mfc='r')
plt.xlabel('Month number')
plt.ylabel('solid units number')
plt.legend('lower right')
plt.grid(axis="x")
plt.show
```

```python
# First Line Graph - Total Profit Over Months
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
axes[0].plot(df["month_number"], df["total_profit"], marker='o', linestyle='-', color='b', label="Total Profit")
axes[0].set_xlabel("Month Number")
axes[0].set_ylabel("Total Profit")
axes[0].set_title("Total Profit Over Months")
axes[0].legend(loc='lower right')
axes[0].grid(True)

# Second Line Graph - Styled Total Profit Over Months
axes[1].plot(df["month_number"], df["total_profit"], marker='o', linestyle='dotted', color='red', 
             markerfacecolor='red', linewidth=3, label="Total Profit (Styled)")
axes[1].set_xlabel("Month Number")
axes[1].set_ylabel("Sold Units Number")
axes[1].set_title("Total Profit Over Months (Styled)")
axes[1].legend(loc="lower right")
axes[1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
plt.Figure(figsize=(10,6))
products=['facecream', 'facewash', 'toothpaste', 'bathingsoap',
       'shampoo', 'moisturizer']
for product in products:
    plt.plot(df['month_number'],df[product],marker='s',linestyle='-',color='g',label=product)
    plt.xlabel("Month Number")
plt.ylabel("Sales Units")
plt.title("Monthly Sales Data for Products")
plt.legend(loc='upper left')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['month_number'],df['toothpaste'],marker='x',color='r')
plt.xlabel=('month')
plt.ylabel('tootpaste production')
plt.legend()
plt.title('Toothpaste Sales')
plt.tight_layout()
plt.grid(True)
plt.show
```

```python
plt.figure(figsize=(10,5))
bar_width = 0.4
plt.bar(df["month_number"] - bar_width/2, df["facecream"], width=bar_width, color='orange', label="Face Cream")
plt.bar(df["month_number"] + bar_width/2, df["facewash"], width=bar_width, color='purple', label="Face Wash")

plt.xlabel("Month Number")
plt.ylabel("Sales Units")
plt.title("Face Cream & Facewash Sales Data (Bar Chart)")
plt.xticks(df["month_number"])
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

```python
plt.figure(figsize=(10,5))
plt.hist(df["total_profit"], bins=5, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('ProfitRange')
plt.ylabel("Frequency")
plt.title("Histogram of Total Profit")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
total_sales = df[products].sum()

plt.figure(figsize=(8,8))
plt.pie(total_sales, labels=products, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'violet', 'orange'])
plt.title("Total Sales Distribution of Products (Pie Chart)")
plt.show()
```
